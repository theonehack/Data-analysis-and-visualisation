```python
import numpy as np
import matplotlib.pyplot as plt
data = {'C':20,'C++':15,'Java':30,'Python':35}
courses = list(data.keys())
values = list(data.values())
fig = plt.figure(figsize = (10, 5))
# creating the bar plot
plt.bar(courses, values, color ='maroon',width = 0.4)
plt.xlabel("Courses offered", fontsize=50)
plt.ylabel("No. of students enrolled", fontsize=25)
plt.title("Students enrolled in different courses", fontsize=5)
plt.show()
```

```python
data = {'C':20,'C++':15,'Java':30,'Python':35}
courses = list(data.keys())
values = list(data.values())
fig = plt.figure(figsize = (10, 5))
# creating the bar plot
plt.bar(courses, values, color ='maroon',width = 0.4)
plt.xlabel("Courses offered", fontsize=10)
plt.ylabel("No. of students enrolled", fontsize=10)
plt.title("Students enrolled in different courses", fontsize=15)
plt.xticks(rotation=90, ha='left')
```

```python
data = {'C':20,'C++':15,'Java':30,'Python':35}
courses = list(data.keys())
values = list(data.values())
fig = plt.figure(figsize = (10, 5))
# creating the bar plot
plt.bar(courses, values, color ='maroon',
width = 0.4)
plt.xlabel("Courses offered", fontsize=10, fontname="Gabriola")
plt.ylabel("No. of students enrolled", fontsize=10)
plt.title("Students enrolled in different courses", fontsize=15, fontname='Franklin Gothic Medium')
plt.xticks(rotation=90, ha='left')
```
